```python
#การสร้างกราฟเส้น
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Bai Jamjuree'
product1=[1.2,2.8,1.3,4.3,3.5]
product2=[2,1.4,3.3,3.1,1.2]
month=[1,2,3,4,5]
plt.title('ยอดขายของผลิตภัณฑ์ใน 5 เดือน',fontname='Bai Jamjuree',size=20)
plt.plot(month,product1,color='g',marker='o',ls='-')
plt.plot(month,product2,color='m',marker='*',ls='-.')
plt.legend(['ชานมไข่มุก','Sunsu'],fontsize='medium',facecolor='pink',edgecolor='red',title='สินค้า')
xlprop = {'size':16,'fontname':'Bai Jamjuree'}
plt.xlabel('เดือน',xlprop)
plt.ylabel('ยอดขาย(ลบ.)',fontname='Bai Jamjuree',size=16)
plt.text(5,5,'นี่คือยอดขายจำลอง')
plt.savefig('testsave.jpg')
#ไม่มีพื้นหลัง plt.savefig('testsave.png',transparent=True)    
plt.show()
```
