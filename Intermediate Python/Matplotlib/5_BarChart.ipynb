```python
#แท่งตั้ง
import matplotlib.pyplot as plt
course=['Python','PHP','Java']
score=[80,50,99]
c=['red','green','yellow']      #สีของแท่ง
plt.bar(course,score,width=0.3,color=c,hatch='/')
plt.xlabel('Course')
plt.ylabel('Score')
plt.title('Score 2021')
plt.show()
```

```python
#แท่งนอน
plt.barh(course,score,color=c,hatch='\/-|')
plt.show()
```

```python
#Stackbar
course=['Python','PHP','Java']
boys=[30,40,20]
girls=[40,50,90]
plt.bar(course,boys,label='Boys',color='blue')
plt.bar(course,girls,label='Girls',color='magenta',bottom=boys)
plt.legend()
plt.title('Student')
plt.show()
```
