```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Bai Jamjuree'
age = [18,17,20,18,20,15,19,20,17,16,15,14,16,17,20]
plt.hist(age)
plt.xlabel('age',size=18)
plt.ylabel('count',size=18)
plt.title('อายุในแต่ละช่วง',size=20)
plt.show()
```

```python
import numpy as np
data=np.random.normal(10,2.5,5000) #สุ่มข้อมูล 5000 ชุดมีค่าเฉลี่ย 10 และส่วนเบี่ยงเบน 2.5
plt.hist(data)          #ไม่กำหนดช่วงข้อมูล (default=10)
plt.hist(data,bins=30)  #กำหนดช่วงเป็น 30 กลุ่ม
plt.show()
```
